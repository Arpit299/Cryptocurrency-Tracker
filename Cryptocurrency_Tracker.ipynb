# Cryptocurrency Tracker
This notebook tracks real-time cryptocurrency data using the CoinGecko API and provides visual analysis.

```python
import requests
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

def get_crypto_data():
    url = 'https://api.coingecko.com/api/v3/coins/markets'
    params = {
        'vs_currency': 'usd',
        'order': 'market_cap_desc',
        'per_page': 10,
        'page': 1,
        'sparkline': False
    }
    try:
        response = requests.get(url, params=params)
        if response.status_code == 200:
            return response.json()
        else:
            print(f'Error: Received status code {response.status_code}')
            return None
    except Exception as e:
        print(f'An error occurred: {e}')
        return None

data = get_crypto_data()
if data:
    df = pd.DataFrame(data)[['name', 'symbol', 'current_price', 'market_cap', 'price_change_percentage_24h']]
    df.columns = ['Name', 'Symbol', 'Price (USD)', 'Market Cap', '24h Change (%)']
    print(f'Last updated: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
    display(df)
else:
    print('Failed to retrieve fresh data.')
```

### Price Comparison Graph
The following bar chart visualizes the current price of the top 10 cryptocurrencies in USD.

```python
if 'df' in locals():
    plt.figure(figsize=(12, 6))
    plot_df = df.sort_values('Price (USD)', ascending=False)
    sns.barplot(x='Price (USD)', y='Name', data=plot_df, hue='Name', palette='viridis', legend=False)
    plt.xscale('log')
    plt.title('Top 10 Cryptocurrencies by Price (Log Scale)')
    plt.xlabel('Price (USD) - Log Scale')
    plt.ylabel('Cryptocurrency')
    plt.grid(True, which='both', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
else:
    print('DataFrame not found.')
```

### 24h Price Change Analysis
This chart shows the percentage change in price over the last 24 hours, highlighting gains and losses.

```python
if 'df' in locals():
    plt.figure(figsize=(12, 6))
    change_df = df.sort_values('24h Change (%)', ascending=False)
    colors = ['green' if x > 0 else 'red' for x in change_df['24h Change (%)']]
    sns.barplot(x='24h Change (%)', y='Name', data=change_df, palette=colors, hue='Name', legend=False)
    plt.title('24h Price Change Percentage (%)')
    plt.xlabel('Change (%)')
    plt.ylabel('Cryptocurrency')
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
else:
    print('DataFrame not found.')
```
